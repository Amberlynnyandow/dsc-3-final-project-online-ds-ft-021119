## Machine Learning 
**Following the no free lunch theorem, there is not one approach that suits all problems, we will try several different ML models to try and predict whether personality traits, age, gender and level of education have an impact on an individuals drug use** 

### Importing Libraries 

```python
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score, KFold, GridSearchCV
from sklearn import tree 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
import warnings 
warnings.filterwarnings('ignore')
```

```python
target = df['Users'] # Creating predictors and target variable 
df.drop(['Amphet', 'Amyl', 'Benzos', 'Cannabis', 'Coke', 'Crack', 'Ecstasy', 'Heroin',
         'Ketamine', 'Legalh', 'LSD', 'Meth', 'Shrooms', 'Nicotine', 'VSA', 'Users'], axis=1, inplace=True)
```

```python
df.head() #checking
```

### Split data into train and test sets 
**We need to have 2 sets of data. One that we use to train the model and one that we use to test the model. Having a test set, that the algorithm is not familar with will exclude any memorization. This way we can be sure our model will perform well when new data is introduced.**

```python
X_train, X_test, y_train, y_test = train_test_split(
    df, target, test_size=0.3, random_state=5) 
```

### k- Fold Cross validation 
**Cross-validation is a resampling procedure used to evaluate machine learning models. The data is shuffled and split into k groups with each group splitting into a test and train set. It's a common statistical approach to observe many results and then take the average.** 

```python
crossvalidation = KFold(n_splits=8, shuffle=True, random_state=1)
```

### kNN Classifier 
**A simple classifier suited for smaller, scaled datasets.**

```python
knn = KNeighborsClassifier() #Instantiate classifier 
knn.fit(X_train, y_train) #Fit classifier to training data
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

### Improving kNN model
**kNN is sensitive to outliers so it's best to keep neighborhoods small. Here, I run a loop to test for the best *k* value. As *k* increases more points are considered when making predictions**

```python
for k in [1, 5, 10, 25, 50, 100, 200, 300]: #looking for best k parameter
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    print('for k = %3i accuracy is %.3f' % (k, knn.score(X_test, y_test)))
```

```python
knn = KNeighborsClassifier(n_neighbors=200) #Instantiate classifier with new parameters 
knn.fit(X_train, y_train) #Fit classifier to training data
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

### Evaluating kNN Classifier 

```python
pred = knn.predict(X_test) #Generating predictions on our test data 
score = f1_score(pred, y_test) #F1 score for model df
knn_test_acc = accuracy_score(pred, y_test) #Accuracy score 
print('Accuracy Score :', knn_test_acc)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))

# store accuracy in a new dataframe
knn_eval = ['kNN Classification', score, knn_test_acc]
models = pd.DataFrame([knn_eval])
```

**Our kNN model has a high accuracy(# of correct predictions) but when it comes to precision(the % a class prediction was right) and recall(% of correct guesses among a class) our model fails miserably by not being able to correctly predict any non-users. There is a strong class imbalanced with the data. Only 20% of participants are classified as non-users. Let's use SMOTE to help balance the data**

```python
from imblearn.over_sampling import SMOTE, ADASYN 

X_resampled, y_resampled = SMOTE().fit_sample(df, target) #adding more users/non-users
print(pd.Series(y_resampled).value_counts())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, random_state=0)
```

### Decision Tree Classifier 
**Decision trees are easy to interpret and visualize. They can easily capture non-linear patterns and there is no need to normalize the data. However, DTs do have a tendency to overfit noisy data and they are biased with imbalanced datasets. So, we will continue to use our SMOTE data. To begin, its prudent to test for the best depth parameter when using a tree classifier. Let's also use cross validation to make sure.**

```python
crossvalidation = KFold(n_splits=5, shuffle=True, random_state=1) 

for depth in range(1, 12):
    tree_classifier = tree.DecisionTreeClassifier(max_depth=depth, random_state=0) #creating classifier object
    if tree_classifier.fit(X_train, y_train).tree_.max_depth < depth: #training 
        break
    score = np.mean(cross_val_score(tree_classifier, X_test, y_test, scoring='accuracy', cv=crossvalidation))
    
    print('Depth: %i Accuracy: %.3f' % (depth, score))
```

### Tuning DT Parameters
**A depth of 9 shows the most promise. By default the model uses the Gini index criterion for splitting. In other words the model wants to decrease the entropy of a dataset by optimizing each split so as to increase the information gained for each split. Gini and entropy criterions performed similarly. Finally, we can allow the model to default to the best splitter method because the random splitter descreased our performance** 

```python
tree_classifier = DecisionTreeClassifier(max_depth=9, criterion="entropy", splitter="best", random_state=5)
tree_classifier = tree_classifier.fit(X_train, y_train)
tree_pred = tree_classifier.predict(X_test)


score = f1_score(tree_pred, y_test) #F1 score for model df
dt_test_acc = accuracy_score(tree_pred, y_test) #Accuracy score 


print('Accuracy score:', dt_test_acc)
print(confusion_matrix(y_test, tree_pred))
print(classification_report(y_test, tree_pred))

dt_eval = ['dt_Classifier', score, dt_test_acc]
models = models.append([dt_eval])
```
